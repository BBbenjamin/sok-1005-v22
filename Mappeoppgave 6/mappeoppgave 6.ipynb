```python
from matplotlib import pyplot as plt

fig,ax=plt.subplots()

ax.set_ylabel('Snitt minutter per kamp')
ax.set_xlabel('Snitt poeng per kamp')

ax.scatter(df2021['PPG'], df2021['MPG'], label = 'Spillere')
ax.legend(loc='lower right', frameon=True)
```

```python
df2021['size']=df2021['GP'].apply(lambda pop: pop)
df2021
```

```python
from bokeh.plotting import figure

#creating figure:
plot = figure(
        title = "2021-2022 Los Angeles Lakers Stats", 
        x_axis_label = 'Snitt poeng per kamp',
        y_axis_label = 'Snitt minutter per kamp',
    

        tools="hover", 
        tooltips = [
            ("Spiller","@Player"),
            ("Snitt poeng","@PPG"),
            ("Snitt minutter","@MPG"),
            ("Kamper spilt","@GP")
            ],
    
        plot_height = 720,
        plot_width = 1280)
```

```python
from bokeh.io import show, output_notebook
from bokeh.plotting import output_file

plot.scatter(
    source=df2021,
    x= 'PPG', 
    y= 'MPG',
    
    size='size',
    
    legend_field='Team',
    alpha= 1
    )

#Creating the graph and saving as html
output_notebook()
output_file("lakers-players.html")
show(plot)
```

```python
plot.legend.location = "top_left"

plot.legend[0].border_line_alpha=0
plot.outline_line_alpha=0
plot.grid[0].grid_line_alpha=0
plot.grid[1].grid_line_alpha=0

output_notebook()
output_file("lakers-players.html")
show(plot)
```
